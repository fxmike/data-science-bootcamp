```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = sns.load_dataset('tips')
df.head(10)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
sns.relplot(data=df, x='total_bill', y='tip');
```

```python
sns.set(font_scale=1.2)
sns.relplot(data=df, x='total_bill', y='tip');
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size');
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time');
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', row='time');
```

```python
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', col='time', row='smoker');
```

```python
sns.catplot(data=df, x='day', y='total_bill');
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='swarm');
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='box');
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='violin');
```

```python
sns.catplot(data=df, x='day', y='total_bill', kind='bar');
```

```python
sns.catplot(data=df, x='day', kind='count');
```
